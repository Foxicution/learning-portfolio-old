```python
import pandas as pd
from midterm.definitions import DATA_DIR

df = pd.read_excel(DATA_DIR / "Midterm data.xlsx")
df.dropna(inplace=True)
df.drop(index=28, inplace=True)
df
```

```python
distribution_place_of_birth = df.groupby(["Gimtasis miestas"])["Gimtasis miestas"].count()
distribution_living_location = df.groupby(["MIestas, kuriame gyvena"])["MIestas, kuriame gyvena"].count()
distribution_place_of_birth
```

```python
distribution = pd.DataFrame({"Gimtasis miestas": distribution_place_of_birth, "Miestas, kuriame gyvena": distribution_living_location})
ax = distribution.plot.bar(rot=70)
ax.set_xlabel("Miestas")
ax.set_ylabel("Skaičius")
```

```python
from matplotlib.ticker import MaxNLocator

hw_counts = df.groupby(["Apsigintų namų darbų kiekis"])["Apsigintų namų darbų kiekis"].count()
ax = hw_counts.plot.bar()
ax.set_xlabel("Apsigintų namų darbų kiekis")
ax.set_ylabel("Skaičius")
```

```python
ax = df["Praeito semestro pažymių vidurkis"].plot.hist(bins=10)
ax.set_xlabel("Praeito semestro pažymių vidurkis")
ax.set_ylabel("Dažnis")
```

```python
ax = df.groupby(["MIestas, kuriame gyvena"])["Praeito semestro pažymių vidurkis"].mean().plot.bar(rot=70)
ax.set_xlabel("Miestas, kuriame gyvena")
ax.set_ylabel("Praeito semestro vidutinis pažymių vidurkis")
```

```python
fig, ax = plt.subplots()
for i in new_df.color.unique():
    df_color = new_df[new_df.color==i]
    df_color.plot.scatter(x='Praeito semestro pažymių vidurkis',y='Apsigintų namų darbų kiekis', ax=ax, c=i, label=df_color["MIestas, kuriame gyvena"].iloc[0])

ax.legend(loc='best')
```
